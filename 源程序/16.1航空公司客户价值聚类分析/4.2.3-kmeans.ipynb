代码4-1 基于K-Means算法的航空公司客户价值细分的实现

```python
import pandas as pd
from sklearn.cluster import KMeans#导入K均值聚类算法
import os
```

```python
filepath=os.getcwd()#获取当前路径
inputfile=filepath+'/customerdata.xlsx'#当前路径下的客户数据
datanew=pd.read_excel(inputfile)
datanew=(datanew-datanew.mean(axis=0))/(datanew.std(axis=0))#对数据进行标准化操作
k=5#需要进行的聚类类别数
#调用K均值聚类算法，进行聚类分析
kmodel=KMeans(n_clusters=k)
kmodel.fit(datanew)
```

```python
datanew['type']=kmodel.labels_#将聚类结果合并入原始数据中，以便查看用户类型
print(datanew )#查看用户聚类结果
```

```python
print(kmodel.cluster_centers_ )#查看各聚类中心
```

对聚类分析的结果进行可视化处理

```python
#导入可视化需要的库
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
```

```python
labels = np.array(['C','F','L','M','R'])#设置标签
dataLength = 5#设置数据长度
angles = np.linspace(0, 2*np.pi, dataLength, endpoint=False)#分割圆周长
angles = np.concatenate((angles, [angles[0]]))  # 闭合
fig = plt.figure(figsize=(5,5))#设置画布大小
ax = fig.add_subplot(111, polar=True)
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")#添加标签
ax.set_rlim(-2,2)
color=['ro-','bs--','gp:','y>-.','k*-']
for k in range(5):
    data=kmodel.cluster_centers_[k]
    data=np.concatenate((data, [data[0]]))#闭合
    ax.plot(angles,data,color[k],linewidth=1)
plt.title(u"五类用户特征雷达图",fontproperties="SimHei",fontsize=30)
plt.show()
```
